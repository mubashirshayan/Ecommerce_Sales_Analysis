# Top 5 Best Selling Products

```python
df2=df.groupby('product')['total'].sum().sort_values(ascending=False).head(5)
```

```python
df2.values
```

```python
df2.index
```

```python
productData={"Product":df2.index,"TotalSales":df2.values}
productData1=pd.DataFrame(productData)
```

```python
productData1
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=productData1,x='Product',y='TotalSales', hue="Product", palette="viridis")
plt.title("Top 5 Best Selling Products")
plt.legend(title="Product")
plt.show()
```

Observations

Headphones are the top-selling product, generating the highest sales at 9.41 million.

Watches are a close second, with sales of 9.30 million, showing strong demand.

Mobiles and Shoes follow, both performing well with over 8.3 million in sales.

Laptops rank fifth, still strong with 7.9 million in sales.

Overall, all top 5 products have high sales, with only a moderate gap between them, indicating broad customer interest across multiple categories.

# Payment Method Distribution

```python
payment_method = df['payment_method'].value_counts()
payment_method
```

```python
plt.figure(figsize=(10,6))
payment_method.plot.pie(autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

Observations

JazzCash is the most used payment method with 112 transactions.

Credit Card and Easypaisa are also popular, with 107 and 105 transactions respectively—almost the same usage.

COD (Cash on Delivery) is moderately used with 97 transactions.

Debit Card is the least used payment method at 79 transactions.

Overall, digital wallets (JazzCash & Easypaisa) and cards dominate, while COD and Debit Card are less preferred.

# Monthwise Sales

```python
month= df.groupby("Month")['total'].sum()
month
```

```python
plt.figure(figsize=(10,5))
plt.plot(month.index,month.values,marker="o")
plt.grid()
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

March recorded the highest sales at 6.71 million, followed closely by April and January, both above 6.2 million.

February and May also performed well, each generating over 5.5 million and 5.9 million.

The lowest sales occurred in September at 2.52 million, with August, November, and July also showing lower performance (between 2.5–2.8 million).

Sales trend shows peaks in the first quarter (Jan–Apr) and a decline from July to November.

Overall, sales are uneven throughout the year, with strong performance in early months and lower activity in mid to late months.



# Show City wise sales

```python
df3=df.groupby('City')['total'].sum().sort_values(ascending=False).head(5)
df3
```

```python
city_wise={"City":df3.index,"Sales":df3.values}
city_wise1=pd.DataFrame(city_wise)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=city_wise1,y="City",x=df3.values,hue="City",palette='Set1')
plt.title("City Wise Sales")
plt.show()
```

Lahore leads by a large margin, recording the highest sales at 14.66 million, significantly outperforming all other cities.

Quetta and Islamabad show comparable performance, with sales of 9.49 million and 9.47 million respectively—indicating strong and nearly equal market activity.

Faisalabad and Peshawar lag behind the top three, generating 6.32 million and 6.19 million in sales.

There is a clear sales gap between the top city (Lahore) and the others—Lahore exceeds the second-highest (Quetta) by more than 5 million.

The bottom two cities (Faisalabad and Peshawar) have similar sales levels, suggesting evenly distributed market potential but lower overall demand compared to the top three.

Overall, sales distribution is uneven, with one dominant city (Lahore), two mid-tier performers (Quetta, Islamabad), and two lower-tier markets (Faisalabad, Peshawar).

# Show monthly Sales by Category

```python
df
```

```python
df4 = df.groupby(["Month","category"])['total'].sum()
```

```python
df4
```

```python
plt.figure(figsize=(10,6))

df_pivot = df4.unstack(level='category')

df_pivot.plot(kind='bar', figsize=(10,5))

plt.title("Monthly Sales by Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category")
plt.tight_layout()
plt.show()
```

Observations

Accessories and Fashion dominate sales in most months, with strong peaks in January, March, and April.

Electronics shows strong performance in March, May, and October, indicating periodic high demand.

Fashion leads in February and May, showing the highest category sales during these months.

Lowest overall sales occur in August, September, and October, across all categories.

March and April stand out as strong months, with all three categories performing well.

Accessories perform consistently well, especially in January, March, and April.

Electronics shows more fluctuation, with lower sales in June and August but strong spikes in March and October.

Fashion is highly variable, dipping significantly in July and September but peaking in February and May.

Overall, sales by category vary month to month, with notable seasonal trends showing higher sales in the first quarter (Jan–Apr) and reduced activity in mid-year months (Jul–Sep)

Lahore leads in city-wise sales, while Quetta and Islamabad perform moderately, and Faisalabad and Peshawar lag behind.
